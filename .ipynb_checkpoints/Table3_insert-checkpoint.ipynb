```python
import sqlite3
```

```python
connection = sqlite3.connect("Otkr")
cursor = connection.cursor()
```

```python
with open('EUR.csv') as f:
    a = [line.strip() for line in f ]
a = [i.replace(",",".")for i in a]
for line in a:
    date = line.split("/")
    if len(date[0]) == 1:
        line = "20" + date[2] + "-0" + date[0] + "-"+ date[1]
    else:
        line = "20" + date[2] + "-0" + date[0] + "-"+ date[1]
    str_ = '("'+ str(line.split(";")[0])+ '","EUR","' + str(line.split(";")[1]) +'")'
    print(str_)
    format_str="""insert into Table_3 VALUES {line} """.format(line = str_)
    print(format_str)
    cursor.execute(format_str)
    connection.commit()

with open('USD.csv') as f:
    b = [line.strip() for line in f ]
b = [i.replace(",",".")for i in b]
for line in b:
    date = line.split("/")
    if len(date[0]) == 1:
        line = "20" + date[2] + "-0" + date[0] + "-"+ date[1]
    else:
        line = "20" + date[2] + "-0" + date[0] + "-"+ date[1]
    str_ = '("'+ ((line.split(";")[0]))+ '","USD","' + (line.split(";")[1])+'")'
    print(str_)
    format_str="""insert into Table_3 VALUES {line} """.format(line = str_)
    print(format_str)
    cursor.execute(format_str)
    connection.commit()
    
```
